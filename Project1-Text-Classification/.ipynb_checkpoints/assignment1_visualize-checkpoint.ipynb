```python
import numpy as np
import os
import platform
import pickle
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import seaborn as sns
```

```python
bert_base_df = pd.read_csv("bert-base-all.csv")
```

```python
bert_base_df.head(5)
```

```python
bert_base_all_x = bert_base_df["Step"]
bert_base_all_y = bert_base_df["Uncased Model Accuracy"]
```

```python

# x=np.linspace(-3,3,50)  #在（-1，1）
bert_base_all_x = bert_base_df["Step"]
bert_base_all_y1 = bert_base_df["Training Loss"]
bert_base_all_y2 = bert_base_df["Validation Loss"]

#一张图里化两条或多条线

plt.figure(dpi=100, figsize=(5, 6))
plt.title('Visualize BERT base Loss')
plt.grid(linestyle='-.')
plt.ylim(0,3)
plt.plot(bert_base_all_x,bert_base_all_y1)
plt.plot(bert_base_all_x,bert_base_all_y2)
plt.show()
```

```python
# x=np.linspace(-3,3,50)  #在（-1，1）
bert_base_all_x = bert_base_df["Step"]
bert_base_all_y3 = bert_base_df["Uncased Model Accuracy"]

#一张图里化两条或多条线

plt.figure(dpi=100, figsize=(5, 6))
plt.title('Visualize BERT base Accuracy')
plt.grid(linestyle='-.')
plt.ylim(0.5)
plt.plot(bert_base_all_x,bert_base_all_y3)
plt.show()
```

```python
roberta_base_df = pd.read_csv("roberta-5-class.csv")
```

```python
roberta_base_df.head()
```

```python
roberta_base_all_x = roberta_base_df["Step"]
roberta_base_all_y = roberta_base_df["Roberta Model Accuracy"]
```

```python
roberta_base_all_x = roberta_base_df["Step"]
roberta_base_all_y1 = roberta_base_df["Training Loss"]
roberta_base_all_y2 = roberta_base_df["Validation Loss"]

plt.figure(dpi=100, figsize=(5, 6))
plt.title('Visualize roBERTa base Loss')
plt.grid(linestyle='-.')
plt.ylim(0,3)
plt.legend()
plt.plot(roberta_base_all_x,roberta_base_all_y1)
plt.plot(roberta_base_all_x,roberta_base_all_y2)
plt.show()
```

```python
roberta_base_all_x = roberta_base_df["Step"]
roberta_base_all_y3 = roberta_base_df["Roberta Model Accuracy"]

plt.figure(dpi=100, figsize=(5, 6))
plt.title('Visualize roBERTa base Accuracy')
plt.grid(linestyle='-.')
plt.ylim(0.4,1)
plt.legend()
plt.plot(roberta_base_all_x,roberta_base_all_y3)
plt.show()
```
